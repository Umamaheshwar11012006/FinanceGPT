## Importing Required Libraries

```python
from transformers import pipeline
```

## Loading GPT2 Model

GPT2 is a pretrained language model used for text generation.

```python
pipe = pipeline(
    "text-generation",
    model="gpt2"
)
```

## Generating First AI Response

```python
result = pipe(
    "Indian stock market today",
    max_new_tokens=50
)

print(result[0]["generated_text"])
```

## Testing Finance Related Prompts

```python
prompts = [
    "What is inflation?",
    "Explain mutual funds",
    "What is SIP investment?",
    "What causes stock market crashes?"
]
```

```python
for prompt in prompts:

    result = pipe(
        prompt,
        max_new_tokens=80
    )

    print("\nPROMPT:")
    print(prompt)

    print("\nRESPONSE:")
    print(result[0]["generated_text"])

    print("\n" + "=" * 80)
```

## Observation

GPT2 can generate finance-related responses,
but:
- responses may contain hallucinations
- financial reasoning is weak
- answers are sometimes incomplete

This shows why fine-tuning is important.

## Understanding Generation Parameters

```python
result = pipe(
    "Explain inflation in simple words",
    max_new_tokens=100,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.2
)

print(result[0]["generated_text"])
```

## Understanding Parameters

max_new_tokens:
- controls response length

temperature:
- controls randomness

top_p:
- controls token selection probability

repetition_penalty:
- reduces repeated text generation

## Testing TinyLlama Model

```python
tinyllama_pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
```

```python
result = tinyllama_pipe(
    "Explain stock market in simple words",
    max_new_tokens=100
)

print(result[0]["generated_text"])
```

## Model Comparison
