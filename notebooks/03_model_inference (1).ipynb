#  FinanceGPT - Model Inference

# This notebook contains experiments with Hugging Face text generation pipelines and tokenizer basics.

```python
from transformers import pipeline
```

```python
pipe = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
result = pipe(
    "Indian stock market today",
    max_new_tokens=50
)

print(result[0]["generated_text"])
```

```python
prompts = [
    "Reliance stock analysis",
    "Infosys quarterly earnings",
    "Future of Indian economy"
]

for prompt in prompts:
    result = pipe(prompt, max_new_tokens=30)
    print("\nPROMPT:", prompt)
    print(result[0]["generated_text"])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "Indian stock market"

tokens = tokenizer.tokenize(text)

print(tokens)
```

# Tokenizer Basics

Tokenizer converts text into smaller units called tokens.
AI models process tokens instead of normal sentences.


```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "Indian stock market"

tokens = tokenizer.tokenize(text)

print(tokens)
```

```python
encoded = tokenizer(text)

print(encoded)
```

# Observations

- GPT2 generates text based on prompts
- Different prompts produce different outputs
- Tokenizer converts text into tokens
- max_new_tokens controls output length
- AI models process tokens internally
