## Testing Text Generation Using GPT2

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2"
)

result = pipe(
    "Indian stock market today",
    max_new_tokens=50
)

print(result[0]["generated_text"])
```

## Observation

GPT2 generates human-like text successfully, but:
- responses may not always be accurate
- financial reasoning is limited
- hallucinations can occur

This shows why better models and fine-tuning are important.

## Testing TinyLlama Model

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)

result = pipe(
    "Explain SIP investment in simple terms",
    max_new_tokens=100
)

print(result[0]["generated_text"])
```

## Model Comparison

GPT2:
- older model
- weaker contextual understanding
- limited financial knowledge

TinyLlama:
- newer chat-oriented model
- better response quality
- improved conversational ability
