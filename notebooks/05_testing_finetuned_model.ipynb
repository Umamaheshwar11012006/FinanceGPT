```python
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline
)

from peft import PeftModel
```

```python
base_model = AutoModelForCausalLM.from_pretrained(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)

tokenizer = AutoTokenizer.from_pretrained(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)

print("Base model loaded successfully.")
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
model = PeftModel.from_pretrained(
    base_model,
    "/content/drive/MyDrive/financegpt-finetuned"
)

print("FinanceGPT loaded successfully.")
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

print("Pipeline created successfully.")
```

```python
prompt = """### Instruction:
How do banks make profit?

### Response:
"""

result = pipe(
    prompt,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.7
)

print(result[0]["generated_text"])
```

```python
prompt = "What is SIP investment?"

result = pipe(
    prompt,
    max_new_tokens=100
)

print(result[0]["generated_text"])
```

```python
prompt = "How do banks make profit?"

result = pipe(
    prompt,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.95
)

print(result[0]["generated_text"])
```
