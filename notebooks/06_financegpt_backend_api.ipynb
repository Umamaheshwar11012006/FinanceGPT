```python
from flask import Flask, request, jsonify

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline
)

from peft import PeftModel
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

base_model = AutoModelForCausalLM.from_pretrained(
    model_name
)

tokenizer = AutoTokenizer.from_pretrained(
    model_name
)

model = PeftModel.from_pretrained(
    base_model,
    "/content/drive/MyDrive/financegpt-finetuned"
)

print("FinanceGPT model loaded successfully.")
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

print("Pipeline created successfully.")
```

```python
app = Flask(__name__)

print("Flask app created successfully.")
```

```python
@app.route("/")
def home():
    return "FinanceGPT Backend API is running successfully."
```

```python
@app.route("/chat", methods=["POST"])
def chat():

    data = request.get_json()

    user_message = data["message"]

    prompt = f"""### Instruction:
{user_message}

### Response:
"""

    result = pipe(
        prompt,
        max_new_tokens=30,
        do_sample=True,
        temperature=0.7,
        truncation=True
    )

    response = result[0]["generated_text"]

    return jsonify({
        "response": response
    })
```

```python
app.run(
    host="0.0.0.0",
    port=5000
)
```

```python
import requests

url = "http://127.0.0.1:5000/chat"

data = {
    "message": "What is SIP investment?"
}

response = requests.post(
    url,
    json=data
)

print(response.json())
```

# Simplified Backend API Simulation

```python
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline
)

from peft import PeftModel
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

base_model = AutoModelForCausalLM.from_pretrained(
    model_name
)

tokenizer = AutoTokenizer.from_pretrained(
    model_name
)

model = PeftModel.from_pretrained(
    base_model,
    "/content/drive/MyDrive/financegpt-finetuned"
)

print("FinanceGPT loaded successfully.")
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)
```

```python
def financegpt_api(user_message):

    prompt = f"""### Instruction:
{user_message}

### Response:
"""

    result = pipe(
        prompt,
        max_new_tokens=30,
        do_sample=True,
        temperature=0.7
    )

    return {
        "response": result[0]["generated_text"]
    }
```
