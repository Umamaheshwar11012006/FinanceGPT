```python
import gradio as gr

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline
)

from peft import PeftModel
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

model = AutoModelForCausalLM.from_pretrained(
    model_name
)

tokenizer = AutoTokenizer.from_pretrained(
    model_name
)

print("Base TinyLlama model loaded successfully.")
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

print("Pipeline created successfully.")
```

```python
def financegpt_chat(user_message):

    prompt = f"""### Instruction:
{user_message}

### Response:
"""

    result = pipe(
        prompt,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.7
    )

    generated_text = result[0]["generated_text"]

    return generated_text
```

```python
interface = gr.Interface(
    fn=financegpt_chat,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Ask FinanceGPT a finance question..."
    ),
    outputs="text",
    title="FinanceGPT",
    description="AI-powered finance assistant built using TinyLlama and Gradio."
)
```

```python
interface.launch()
```
